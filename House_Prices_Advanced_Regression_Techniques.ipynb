## **2.7.1 ANOVA Test:**

Earlier we used correlation to determine the influence of numeric features on SalePrice.Similarly, we'll use the ANOVA test to understand the correlation between categorical variables and SalePrice.Why ANOVA?Because the SSE(Sum of squared errors) in the ANOVA table tells you the proportion of variance explained by the feature or groups of features to the total variance in the data. Obviously the features that explain the largest proportion of the variance should be retained.

Now, we'll define a function which calculates p values. From those p values, we'll calculate a disparity score. Higher the disparity score, better the feature in predicting sale price.


```python
cat = [f for f in SystematicAnalysis.columns if SystematicAnalysis.dtypes[f] == 'object']
def anova(frame):
    anv = pd.DataFrame()
    anv['features'] = cat
    pvals = []
    for c in cat:
           samples = []
           for cls in frame[c].unique():
                  s = frame[frame[c] == cls]['SalePrice'].values
                  samples.append(s)
           pval = stats.f_oneway(*samples)[1]
           pvals.append(pval)
    anv['pval'] = pvals
    return anv.sort_values('pval')
pd.options.mode.chained_assignment = None # default='warn'
cat_data_train['SalePrice'] = SystematicAnalysis.SalePrice.values
k = anova(cat_data_train) 
k['disparity'] = np.log(1./k['pval'].values) 
sns.barplot(data=k, x = 'features', y='disparity') 
plt.xticks(rotation=90) 
plt.show()
```

According to this bar chart Neighborhood seems to be the most important feature followed by ExterQual, KitchenQual, etc.This means that people also consider the goodness of the neighborhood, the quality of the kitchen, the quality of the material used on the exterior walls, etc

# **8.  Box Cox Transformation**

```python
data.skew()
```

```python
# Most Skewed Features
NumF = data.dtypes[data.dtypes != 'object'].index


# Check the skew of all numerical features
skewed_feats = data[NumF].apply(lambda x: skew(x.dropna())).sort_values(ascending=False)
print("\nSkew in numerical features: \n")
skewness = pd.DataFrame({'Skew' :skewed_feats})
skewness.head(15)

skewness = skewness[abs(skewness) > 0.75]
print("There are {} skewed numerical features to Box Cox transform".format(skewness.shape[0]))

from scipy.special import boxcox1p
skewed_features = skewness.index
lam = 0.15
for feat in skewed_features:
    #all_data[feat] += 1
    data[feat] = boxcox1p(data[feat], lam)
```

# **Cell Report:**

1. When we deal with real-world data, we need to deal with features that are heavily skewed. Transformation technique is useful to stabilize variance, make the data more normal distribution-like, improve the validity of measures of association.Due to that we have used  Box-Cox Transformation.


2. I have used the Box-Cox transformation to transform non-normal variables into a normal shape. Normality is an important assumption for many statistical techniques; if our data isn't normal, applying a Box-Cox means that we are able to run a broader number of tests.


3. The problem with the Box-Cox Transformation is estimating lambda. Here after several trial & error i found 0.15 as the best lambda value, which worked best for my model.

# **13. Decision of Hyper Parameters**

```python
A machine learning model has two types of parameters. The first type of parameters are the parameters that are learned through a machine learning model while the second type of parameters are the hyper parameter that we pass to the machine learning model. A good choice of hyperparameters can really make an algorithm shine. The process of finding most optimal hyperparameters in machine learning is called hyperparameter optimization. Here I am using Grid search technique for hyperparameter optimization through which we can come to know the best parameters for the model. Randomly selecting the parameters for the algorithm can be exhaustive. Therefore, instead of randomly selecting the values of the parameters, a better approach would be to use Grid Search which automatically finds the best parameters for a particular model. Grid search measures the performance using “Cross Validation” technique. This validation technique gives assurance that our trained model got most of the patterns from the dataset which is the main reason why I have used grid search.
```

```python
train_data_dl_withoutpca=np.array(train_data[0:1397])
train_labels_dl_withoutpca=np.reshape(np.array(train_labels[0:1397]),(1397,1))
Validation_data_dl_withoutpca=np.array(validation_data[0:59])
Validation_labels_dl_withoutpca=np.reshape(np.array(validation_labels[0:59]),(59,1))
```

## **14.7.1 AdamOptimizer:**

## **i) Epoch = 500**

```python
X = tf.placeholder("float", [None, 297])#297 features 
Y = tf.placeholder("float", [None, 1])#1 output label 
weights = {
    'h1': tf.Variable(tf.random_normal([297, 10])),#297 inputs 10  nodes in h1 layer
    'out': tf.Variable(tf.random_normal([10,1]))# 1 ouput label
}
biases = {
    'b1': tf.Variable(tf.random_normal([10])),
    'out': tf.Variable(tf.random_normal([1]))
}
def neural_net(x):
    #hidden layer 1
    layer_1 = tf.add(tf.matmul(x, weights['h1']), biases['b1'])
    layer_1_1 = tf.nn.leaky_relu(layer_1)#activation
    # output layer
    out_layer = tf.matmul(layer_1_1, weights['out']) + biases['out']
    return (out_layer)
Y_hat=neural_net(X)
loss_op=tf.losses.mean_squared_error(Y,Y_hat)
lr_rate=[0.1,0.01,0.2,0.02,0.3,0.03,0.4,0.04,0.5,0.05,0.6,0.06,0.7,0.07,0.8,0.08,0.9,0.09,1.0]
loss_train_withoutpca_Adam_E500=[]
loss_validation_withoutpca_Adam_E500=[]
for z in lr_rate:
    train_op = tf.train.AdamOptimizer(z).minimize(loss_op)
    init = tf.global_variables_initializer()
    epoch=500
    with tf.Session() as sess:
        sess.run(init)
        for i in range(0,epoch+1):
            sess.run(train_op,feed_dict={X:train_data_dl_withoutpca,Y:train_labels_dl_withoutpca})
            loss=sess.run(loss_op,feed_dict={X:train_data_dl_withoutpca,Y:train_labels_dl_withoutpca})
            loss_Validation=sess.run(loss_op,feed_dict={X:Validation_data_dl_withoutpca,Y:Validation_labels_dl_withoutpca})
            if(i==500):
                loss_train_withoutpca_Adam_E500.append(loss**(.5))
                loss_validation_withoutpca_Adam_E500.append(loss_Validation**(.5))
                print("epoch no"+str(i),"Training loss: ", (loss**(.5)),"Validation loss: ", (loss_Validation**(.5)), "Learning Rate: ", (z))
```

```python
n_groups = 19
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.45
opacity = 1

rects1 = plt.bar(index, loss_train_withoutpca_Adam_E500, bar_width,
alpha=opacity,
color='b',
label='Training Loss')

rects2 = plt.bar(index + bar_width, loss_validation_withoutpca_Adam_E500, bar_width,
alpha=opacity,
color='g',
label='Validation Loss')

plt.xlabel('Learning Rate')
plt.ylabel('RMSE')
plt.title('Training Loss & Validation Loss Based on Various Learning Rate(Epoch=500)')        
plt.xticks(index + bar_width, ('LR=0.1', 'LR=0.01', 'LR=0.2', 'LR=0.02','LR=0.3', 'LR=0.03', 'LR=0.4', 'LR=0.04','LR=0.5', 'LR=0.05', 'LR=0.6', 'LR=0.06','LR=0.7','LR=0.07', 'LR=0.8', 'LR=0.08', 'LR=0.9','LR=0.09','LR=1'))
plt.legend()

plt.tight_layout()
plt.show()
```

## **14.7.9  Final Model Based on Deep Learning without PCA:**

```python
X = tf.placeholder("float", [None, 297])#297 features 
Y = tf.placeholder("float", [None, 1])#1 output label 
weights = {
    'h1': tf.Variable(tf.random_normal([297, 10])),#297 inputs 10  nodes in h1 layer
    'out': tf.Variable(tf.random_normal([10,1]))# 1 ouput label
}
biases = {
    'b1': tf.Variable(tf.random_normal([10])),
    'out': tf.Variable(tf.random_normal([1]))
}
def neural_net(x):
    #hidden layer 1
    layer_1 = tf.add(tf.matmul(x, weights['h1']), biases['b1'])
    layer_1_1 = tf.nn.leaky_relu(layer_1)#activation
    # output layer
    out_layer = tf.matmul(layer_1_1, weights['out']) + biases['out']
    return (out_layer)
Y_hat=neural_net(X)
loss_op=tf.losses.mean_squared_error(Y,Y_hat)
optimizer = tf.train.AdamOptimizer(learning_rate=0.4)
train_op = optimizer.minimize(loss_op)
init = tf.global_variables_initializer()
epoch=2500
with tf.Session() as sess:
    sess.run(init)
    for i in range(0,epoch+1):
        sess.run(train_op,feed_dict={X:train_data_dl_withoutpca,Y:train_labels_dl_withoutpca})
        loss=sess.run(loss_op,feed_dict={X:train_data_dl_withoutpca,Y:train_labels_dl_withoutpca})
        loss_Validation=sess.run(loss_op,feed_dict={X:Validation_data_dl_withoutpca,Y:Validation_labels_dl_withoutpca})
        pred=sess.run(Y_hat,feed_dict={X:test_data})
```

## **14.9 Final Deep Learning Model:**

```python
X = tf.placeholder("float", [None, 225])#225 features 
Y = tf.placeholder("float", [None, 1])#1 output label 
weights = {
    'h1': tf.Variable(tf.random_normal([225, 10])),#225 inputs 10  nodes in h1 layer
    'out': tf.Variable(tf.random_normal([10,1]))# 1 ouput label
}
biases = {
    'b1': tf.Variable(tf.random_normal([10])),
    'out': tf.Variable(tf.random_normal([1]))
}
def neural_net(x):
    #hidden layer 1
    layer_1 = tf.add(tf.matmul(x, weights['h1']), biases['b1'])
    layer_1_1 = tf.nn.leaky_relu(layer_1)#activation
    # output layer
    out_layer = tf.matmul(layer_1_1, weights['out']) + biases['out']
    return (out_layer)
Y_hat=neural_net(X)
loss_op=tf.losses.mean_squared_error(Y,Y_hat)
optimizer = tf.train.AdamOptimizer(learning_rate=0.6)
train_op = optimizer.minimize(loss_op)
init = tf.global_variables_initializer()
epoch=3000
with tf.Session() as sess:
    sess.run(init)
    for i in range(0,epoch+1):
        sess.run(train_op,feed_dict={X:train_data_dl_withpca,Y:train_labels_dl_withpca})
        loss=sess.run(loss_op,feed_dict={X:train_data_dl_withpca,Y:train_labels_dl_withpca})
        pred=sess.run(Y_hat,feed_dict={X:test_data_PCA})
```
